1. Select all females from Florida. Expected columns in order:

first_name as FirstName
last_name as LastName
salary as Salary
gender as Gender
state as State

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("population.db")

df = pd.read_sql("SELECT * FROM population", con=conn)

Females_Florida = df[(df["state"] == "Florida") & (df["gender"]=='Female')]

Females_Florida
```
